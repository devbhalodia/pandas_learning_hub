#2D line plot:-


* Just a simple line graph which connects points
* Bivariate analysis (requires 2 variables for the 2 axis)
* those 2 vairables could be cat-num or num-num
* **Use case**:- Used mainly for time series data, where you want to see the trends with time passing. Here, looking at the trend with time, as in, time based progression is much more important then comparing the data at 2 different times. we want to analyse how things change with time, and hence, in case of line plot, it is a requirement for the time to be in continuity, that is, no gaps in between. for eg:- year by year analysis, and not few random years given with gaps and no order.



```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

#plt.plot(x_axis, y_axis)
plt.plot(year,price)
```

```python
df = pd.read_csv('sharma-kohli.csv')
df.head()
```

```python
#we want to plot the kohli's runs, year by year to see his progression:-
plt.plot(df['index'], df['V Kohli'])
```

```python
#now say, you want to check kohli's and sharma's year by year progression side by side:-
'''
you first need to re-run the kohli's graph as it disappears after every run. after, you ran sharma's graph. it basically overlaps with
the kohli's graph. in other words, sharma's graph gets 'PLACED' on top of kohli's graph.
'''
plt.plot(df['index'], df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
```

```python
#giving title to the graph and labels to the axis:-
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#just by looking at the graph, you want come to know which coloured line is for whom, right? hence we use legend.
'''
NOTE:- for legend, we need to label each plot and then only will it get the legend successfully.
'''
plt.plot(df['index'],df['V Kohli'], label = 'Virat')
plt.plot(df['index'],df['RG Sharma'], label = 'Sharma')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
```

```python
#if you want to change the line type for further differentiation, use linestyle:-
plt.plot(df['index'],df['V Kohli'],linestyle='solid', label = 'Virat')
plt.plot(df['index'],df['RG Sharma'],linestyle='dashed', label = 'Sharma')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
```

```python
#if you want to mark the values, as in, number of runs scored each year, use markers:-
plt.plot(df['index'],df['V Kohli'],linestyle='solid',marker = 'D' ,label = 'Virat')
plt.plot(df['index'],df['RG Sharma'],linestyle='dashed',marker = 'o',label = 'Sharma')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
```

```python
'''
when too extreme outliers are present, you cannot check the variations in values between the other points, because the last point
dominated too much.
'''
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year, price)

'''
in order to prevent this from happening, you can eliminate the outliers during plotting itself, by setting a limit for both the axis.
'''
plt.ylim(0,75000)
plt.xlim(2016,2019) #if you only want the data between these 2 years.
```

```python
#say, you want to map the values in the graph, it becomes easier when using grid:-
plt.plot(df['index'],df['V Kohli'],linestyle='solid',marker = 'D' ,label = 'Virat')
plt.plot(df['index'],df['RG Sharma'],linestyle='dashed',marker = 'o',label = 'Sharma')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
plt.grid()
```

#Scatter plots:-

* used when we have num-num columns
* hence, bivariate analysis
* **use case**:- helps visually finding the correlation between 2 numerical columns.

```python
df = pd.read_csv('batter.csv')
df = df.head(50) #just 50 datapoints required as of now.
df.head()
```

```python
'''
avg:- on avg, how much does a batsman score in a match. a good avg, say an avg of 50, does not mean the batsman is good because say,
he does 50 runs in 80, 100, 120 balls, showing how slow he is.
strike rate:- how many runs he could do in 100 balls. but, 100th ball tak pahucha to. varna judging a batsman just by looking at the
strike rate is a bit off, as it might suggest that the gameplay of batsman is risky. if strike rate too high, har ek ball me 4 or 6
lagana hota hai bhaisaab ko.

so, when a selecting batsman for a team, the bade log look for those who have good avg as well as good strike rate, suggesting that
player can play many balls, alongside scoring many runs.
'''


'''
the upper right quarter has what we want. high avg and high strike rate.
the lower left quarter is something no one wants.
'''
plt.scatter(df['avg'], df['strike_rate'])
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
tips = sns.load_dataset('tips')
tips.head()
#we want to check if the amount of tip correlates with the total bill or not:-
plt.scatter(tips['total_bill'], tips['tip']) #and yes, it positively correlates!
plt.title('tips vs total bill')
plt.xlabel('total bill')
plt.ylabel('tips')
```

#Bar chart:-
* Used when comparison matters more than the trend. you just want to compare the data in years or categories. continuity or order does not matter much.
* num-cat
* Hence, bivariate analysis
* **Use case**:- aggregate analysis of categories, for eg:- category wise total sales. here, you just want to compare the total sales across different categories, and not checking any trends

```python
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
#if you want to horizontally plot it.
plt.barh(colors,children)
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'])
```

```python
#now say, i want a side by side comparison of the runs scored by them in 2015 vs 2016
'''
we can't do this here, as the second plot is placed on top of first plot, hence the overlap seen below.
'''
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,label = '2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2, label = '2016')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,label = '2017')
plt.legend()
'''
np.arrange(df.shape[0]) gives us the location of each data point. by doing +- 0.2 to it, we are shifting the position.

we used the below line, xticks, so that the x axis bars gets cricketers' names as labels and not just numbers 1,2,3,4.........
'''
plt.xticks(np.arange(df.shape[0]), df['batsman'])
```

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

#Histogram
* only used during **Univariate analysis** of **numerical** columns
* **use case**:- to convert the numerical columns into categorical columns (by introducing range) and demonstrating frequency count of each range

```python
data = [32,45,56,10,15,27,61]
'''
i want to represent the whole data. lowest = 10 and highest = 61. so i have to set the bins in such a way that all the points
covered. below, i have set the bin size to 15.
'''
#syntax = plt.hist(univar_data, bins = [, , , , ])
plt.hist(data,bins=[10,25,40,55,70])
```

#Pie chart
* Univariate / Bivariate analysis
* if bivariate, the categorical vs numerical
* **use case**:- used to find the % contribution of each category

```python
# Univariate use case:-
marks = [20,30,14,12,10]
plt.pie(marks)
plt.show()
```

```python
#labels can be given to each part:-
'''
NOTE:- when you give the labels, you are doing bivariate analysis as labels will always be your categorical columns and the column
on which you are making your pie chart, here marks, will be your numerical column in case of cat-num bivariate analysis.
'''
subjects = ['eng', 'maths', 'sci', 'env', 'arts']
plt.pie(marks, labels = subjects)

plt.show()
```

```python
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels = df['batsman'])
plt.show()
```

```python
#when you want to see the % of contribution:-
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```
