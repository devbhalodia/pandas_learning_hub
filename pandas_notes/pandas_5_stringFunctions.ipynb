```python
# syntax = s['col_name'].str.operation()
df = pd.read_csv('titanic (1).csv')
df.head(3)
```

```python
#upper/lower
#df['Name'].str.upper()
df['Name'].str.lower()
```

```python
#capitalize, title, len
df['Name'].str.capitalize() #just the 1st char of the 1st word of entire string ---> capital
df['Name'].str.title() #1st char of all the words of entire string ---> capital
df['Name'].str.len() #length of strings.
```

```python
#strip / split
df['Name'].str.strip() #removes any leading or trailing whitespaces.
df['Name'].str.split() #when nothing mentioned, gives us list of strings separated by whitespaces.
df['Name'].str.split(',') #splits the strings on the basis of comma.

'''
we want to have title and first name of each person. so, firstly, split on the basis of ',' so that the lastname is separated from the
full name. lastname is at index 0 of list which we don't need. we need the second part at index one which has title and firstname. then
we split on the basis of whitespace, but with n=1, which means just split on the basis of first whitespace and keep the rest intact. so
the string just split into 2 parts, 1 before 1st whitespace and other after the 1st whitespace.
MOST IMPORTANTLY, we have used expand=True which will split the splitted string into 2 different columns, creating a df.
'''
df[['title','firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True)
df.head()
```

```python
#replace ('old_value', 'new_value')
df['title'] = df['title'].str.replace('Miss', 'Ms')
```

```python
#filtering
#startswith/endswith
df[df['firstname'].str.startswith('M')]
df[df['firstname'].str.endswith('e')]
```

```python
#isdigit/isalpha
'''
isdigit ---> returns boolean true for all the strings which are completely made up of numbers like '1234'
isalpha ---> returns boolean true for all the strings which are completely made up of alphabets like 'ksadd'
'''
df[df['title'].str.isdigit()]
df[df['title'].str.isalpha()]
```

```python
#contains ---> another word for regex
df[df['firstname'].str.contains('john',case=False)] #returns all firstname starting with john, irrespective of upper/lower case letters.

# find firstname with start and end char vowel
df[df['firstname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```

```python
# slicing (getting substrings or subsequences of the strings in a column/series)
df['Name'].str[::-1]
```
