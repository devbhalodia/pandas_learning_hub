```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('expense_data.csv')
df.head()
```

#obj to datetime

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

#dt accessor (the date part of the timestamp)

```python
df['Date'].dt.date #gives just the date part of the timestamp
df['Date'].dt.year #gives year of the date in 4 digit number (2025)
df['Date'].dt.month #gives number of the month (7)
df['Date'].dt.day #gives number of the day (27)
df['Date'].dt.weekday #gives the day of the week in number (for eg:- 3)
df['Date'].dt.dayofyear #gives the nth day out of the 365 days, for eg:- 207
df['Date'].dt.week #gives the week number going on out of 52 weeks, for eg:- 27
df['Date'].dt.quarter #gives which quarter of the year we are on, for eg:- 3
```

#dt accessor (the time part of the timestamp)

```python
#please note that having the date and its parts, that is, hour, minute, seconds is necessary to have. otherwise, default value 0 returned
df['Date'].dt.time #gives the time part of the timestamp
df['Date'].dt.hour #gives the hour in 24 hour format (16)
df['Date'].dt.minute #gives minute in range 0-59, like 12
df['Date'].dt.second #gives second in range 0-59, like 16
df['Date'].dt.microsecond #12345
```
