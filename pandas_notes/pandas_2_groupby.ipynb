```python
#len
#gives us total number of genres/groups
len(movies.groupby('Genre'))
```

```python
#size
#gives us total number of items in each genre.
movies.groupby('Genre').size()
```

```python
#first, last, or nth item in the group
genres = movies.groupby('Genre')
genres.first() #gives first item/row of each group, not necessarily the smallest or biggest as not ordered.
genres.last() #gives last item/row of each group
genres.nth(6) #7th item/row of each group
```

```python
#get_group
genres.get_group('Fantasy')
'''
each group in itself is a df. so, when we have say the original df being divided into 14 different genres, each genre group is
also a df. hence, Fantasy group is a df.
'''
```

```python
#describe()
'''
gives all the mathematical calculations for each group, be it min, max, q1, q3, median, mean, etc.
'''
genres.describe()
```

```python
#sample
#gives n random samples.
#replace=True because a genre may sample size < 2, and hence if sample size = 1 and we demanded 2 samples, we will get a copy of that one sample.
genres.sample(2,replace=True)
```

```python
'''
counts unique items in each genre. it is similar to size().
size() just returns total size of each group, without eliminating the non unique occurences.
whereas, nunique counts the unique number of items in each group and for each column of that group.
'''
genres.nunique()
```

#aggregate methods ---> .agg()

```python
#agg() can have a dictionary or a list inside

#passing dict:-  agg({})
'''
when you want to have different aggregate functions being applied on different columns, but at the same time, we use dictionary.
'''
genres.agg({
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'min'
})
```

```python
#passing list:- agg([])
'''
when passing agg in form of list, it applies all the mentioned functions to each column. so if some column is non-nnumeric, it will
show errors in case of mean. so, firstly select all the numeric columns and then apply agg.
this is not like passing dict in agg, where you can select what function to apply where.
'''
movies.select_dtypes(include='number').groupby(movies['Genre']).agg(['min', 'max', 'mean'])
```

```python
#using both, the list as well as the dict:-
genres.agg({
    'Runtime':['mean', 'sum'],
    'IMDB_Rating':'mean',
    'No_of_Votes':['min','sum'],
    'Gross':'sum',
    'Metascore':'min'
})
```

#looping in groups

```python
#find the movie details where imdb rating of that movie is the max in each group.
'''
for group,df in genres: ---> this means that we are going to each group's df.
in that df, we need to select those rows where the imdb rating = max imdb rating of the whole column/series.
df[WHERE]
df[df['IMDB_Rating'] == df['IMDB_Rating'].max()]
this will give us row/rows of movie details in the form of dataframe that we append in the list.
at last, we simply concatenate the whole list into the new dataframe, which is, new df.
'''
new_df_list = []
for group,df in genres:
  new_df_list.append(df[df['IMDB_Rating'] == df['IMDB_Rating'].max()])

new_df = pd.concat(new_df_list)
new_df
```

#apply after groupby

---
1. "**split**" using groupby
2. "**apply**" a function
3. "**combine**" the resultant to the original dataframe



```python
# find number of movies starting with A for each group

def check (group):
  cnt = group['Series_Title'].str.startswith('A').sum()
  return cnt

genres.apply(check)
```

```python
# find ranking of each movie in the group according to IMDB score
def give_rank(group):
  group['rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group

genres.apply(give_rank)
```

```python
# find normalized IMDB rating group wise
'''
normalized value = (X-Xmin)/(Xmax - Xmin)
'''
def normalize(group):
  group['normalized_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normalize)
```
